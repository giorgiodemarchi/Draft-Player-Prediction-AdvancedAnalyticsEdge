### Merged dataset

```python
df = pd.merge(players_df, first_round_drafted, on=['player_name','year','team'], how='left')
df.head()
```

```python
df['first_round_picked'] = np.where(df['round'].isnull(), 0, 1)
```

```python
df.info()
```

**First Model**

```python
df = df.drop(['Unnamed: 64', 'Unnamed: 65', 'affiliation', 'round',
       'round_pick', 'overall_pick'], axis=1)
```

```python
# Drop columns with more than 100 missing values
df = df.dropna(thresh=len(df) - 100, axis=1)

# Python
df = df.select_dtypes(include=['int64','int32', 'float64'])
```
